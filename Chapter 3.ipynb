### 3.1.2 數據類型和運算符

```python
x = 10
y = 3
print(x + y)  # 輸出: 13 (加法)
print(x - y)  # 輸出: 7 (減法)
print(x * y)  # 輸出: 30 (乘法)
print(x / y)  # 輸出: 3.3333333333333335 (除法)
print(x // y) # 輸出: 3 (整除)
print(x % y)  # 輸出: 1 (取模)
```

## 3.2 Python常用數據結構

### 3.2.1 列表（List）

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  # 輸出: "apple"
print(fruits[1:3])  # 輸出: ["banana", "cherry"]
```

```python
fruits = ["apple", "banana", "cherry", "orange", "grape"]
```

```python
print(fruits[1:4])  # 輸出: ['banana', 'cherry', 'orange']
```

### 3.2.2 元組（Tuple）

```python
birthday = (6, 15)
```

```python
print(birthday[0])  # 輸出: 6
print(birthday[1])  # 輸出: 15
```

### 3.2.3 集合（Set）

```python
collection = {"stamp", "coin", "painting", "sculpture", "stamp"}
```

```python
my_collection = {"stamp", "coin", "painting"}
friend_collection = {"coin", "sculpture", "vase"}
print(my_collection & friend_collection)  # 輸出: {"coin"}
print(my_collection - friend_collection)  # 輸出: {"stamp", "painting"}
```

### 3.2.4 字典（Dictionary）

```python
person_info = {
"name": "Alice",
"age": 25,
"city": "New York"
}
```

```python
print(person_info["name"])  # 輸出: "Alice"
print(person_info["age"])   # 輸出: 25
print(person_info["city"])   # 輸出: "New York"
```

```python
person_info["age"] = 26
print(person_info["age"])   # 輸出: 26
```

## 3.3 Python控制流程

#### 3.3.1 條件語句

```python
score = 85
if score >= 90:
	print("A")
elif score >= 80:
	print("B")
else:
	print("C")
```

#### 3.3.2 循環語句

```python
prices = [5.99, 12.49, 3.75, 9.99, 7.25]
total = 0
for price in prices:
	total += price
print("Total:", total)
```

```python
count = 0
while count < 5:
	print("Do pushup", count + 1)
	count += 1
print("Great job!")
```

## 3.4 Python函數和模塊

#### 3.4.1 函數（Function）

```python
def calculate_bmi(height, weight):
	bmi = weight / (height ** 2)
	return bmi

print(calculate_bmi(1.75, 70))  # 輸出: 22.857142857142858
```

```python
def send_birthday_wishes(name):
	return f"Happy birthday, {name}!"

print(send_birthday_wishes("Alice"))  # 輸出: "Happy birthday, Alice!"
```

#### 3.4.2 模塊（Module）

```python
import math
print(math.pi)  # 輸出: 3.141592653589793
```
